Part-I

```python
import pandas as pd
import numpy as np
dataset=pd.read_csv('AWCustomers.csv')
dataset.head()
```

```python
print(dataset.info())
```

```python
selected_columns = ['CustomerID','Education', 'Occupation', 'Gender', 'MaritalStatus','HomeOwnerFlag',
                    'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
dataset=dataset[selected_columns]
print(dataset.info())
```

```python
dataset.head()
```

Part-II

```python
print(dataset.isnull())
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
ratio_columns = ['NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
scaler = MinMaxScaler()
dataset[ratio_columns] = scaler.fit_transform(dataset[ratio_columns])
dataset.head()
```

```python
bins = 3
labels = ['Low', 'Medium', 'High']
dataset['YearlyIncome'] = pd.cut(dataset['YearlyIncome'], bins=bins, labels=labels)
dataset['NumberCarsOwned'] = pd.cut(dataset['NumberCarsOwned'], bins=bins, labels=labels)
dataset['NumberChildrenAtHome'] = pd.cut(dataset['NumberChildrenAtHome'], bins=bins, labels=labels)
dataset['TotalChildren'] = pd.cut(dataset['TotalChildren'], bins=bins, labels=labels)
print(dataset.head())
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
cat_columns = ['Education', 'Occupation', 'Gender', 'MaritalStatus',
               'NumberCarsOwned', 'NumberChildrenAtHome', 'TotalChildren', 'YearlyIncome']
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), cat_columns)], remainder='passthrough')
data_encoded = ct.fit_transform(dataset)
encoded_columns = ct.get_feature_names_out()
dataset = pd.DataFrame(data_encoded, columns=encoded_columns)
print(dataset.head())
```

```python
dataset.head()
```

```python
from scipy.spatial.distance import cosine
from sklearn.metrics import jaccard_score

dfa_encoded_numeric = dataset.apply(pd.to_numeric, errors='coerce')

dfa_encoded_numeric = dataset.fillna(0)

row1=dfa_encoded_numeric.iloc[0]
row2=dfa_encoded_numeric.iloc[1]
binary_row1=(row1 > 0).astype(int)
binary_row2=(row2 > 0).astype(int)

binary_row1_array = binary_row1.to_numpy()
binary_row2_array = binary_row2.to_numpy()

jacc_score = jaccard_score(binary_row1_array, binary_row2_array,average='binary')
cosine_similarity=1-cosine(binary_row1_array,binary_row2_array)
cosine_similarity
```
